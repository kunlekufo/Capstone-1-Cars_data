```python
# Reload the CSV file with the correct separator
data = pd.read_csv('C:\\Users\\Kunle Kuforiji\\OneDrive\\Documents\\cars.csv', sep=';')

# Display the first few rows of the DataFrame again
data.head()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Remove the first row that contains data type information
data_cleaned = data.drop(index=0)

# Convert numerical columns to the correct data type
numerical_columns = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model']
data_cleaned[numerical_columns] = data_cleaned[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Check the cleaned data
data_cleaned.head()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['MPG'], kde=True, bins=30)
plt.title('Distribution of Miles Per Gallon (MPG)')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='Weight', y='Horsepower', hue='Origin', style='Origin', palette='deep')
plt.title('Horsepower vs. Weight by Origin')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.legend(title='Origin')
plt.show()

```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=data_cleaned, x='Cylinders', y='MPG', ci=None, palette='coolwarm')
plt.title('Average MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.show()
```

```python
# Comparison of Car Origins
plt.figure(figsize=(10, 6))
sns.countplot(data=data_cleaned, x='Origin', palette='pastel')
plt.title('Comparison of Car Origins')
plt.xlabel('Origin')
plt.ylabel('Count')
plt.show()
```
